```python
# we will check stationarity using Dicky Fuller test
```

```python
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
```

```python
def check_adf(series,check=0.05):
    adf=adfuller(series,autolag='AIC')
    
    print(f"H0: {series.name} is not sationary")
    print(f"H1: {series.name} is stationary")
    
    test_Stat=adf[0];print(f"ADF test statistics: {adf[0]}")
    pvalue=adf[1];print(f"Pvalue is: {adf[1]}")
    print(f"No of logs: {adf[2]}")
    print("\n Critical Values \n")
    for key, items in adf[4].items():
        print("\t",key,"\t",items)
        
    print(f" \n For {check} significance level: \n")
    if pvalue<check:
        print("We can reject null hypothesis. Series is Stationary")
    else:
        print("We can not reject null hypothesis. Series is not stationary")
            
        
```

```python
print("Performance of Augmented Dicky Fuller Test:")
check_adf(price_df_daily['TotalRevenue'])

print("\nPerformance of Augmented Dicky Fuller Test on work day:")
check_adf(dailysales_workday['TotalRevenue'])
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn.metrics import r2_score
```

```python
result=seasonal_decompose(dailysales_workday['TotalRevenue'],model='multiplicative',period=30)
fig=plt.figure(figsize=(15,10))
fig=result.plot()
fig.set_size_inches(16,9)
```

```python
train_size=int(0.8*len(dailysales_workday))
```

```python
train=dailysales_workday.iloc[:train_size]
test=dailysales_workday.iloc[train_size:]
```

```python
train.shape,test.shape
```

```python
train=train['TotalRevenue'].values
test=test['TotalRevenue'].values
history = [x for x in train]
model_predictions = []
N_test_observations = len(test)
for time_point in range(N_test_observations):
    model = ARIMA(history, order=(4,1,0))
    model_fit = model.fit(disp=0)
    output = model_fit.forecast()
    yhat = output[0]
    model_predictions.append(yhat)
    true_test_value = test[time_point]
    history.append(true_test_value)
MSE_error = mean_squared_error(test, model_predictions)
print('Testing Mean Squared Error is {}'.format(MSE_error))
MAE_error=mean_absolute_error(test,model_predictions)
print('Testing Mean Absolute Error is {}' .format(MAE_error))
r2_score=r2_score(test,model_predictions)
print('R2 score is {}'.format(r2_score))
```
